Crear una matriz de Numpy aleatoria de 4 dimensiones y un size de 1200000

```python
import numpy as np

matriz_4d = np.random.random(size=(10, 20, 30, 200))
print("Shape:", matriz_4d.shape)
print("Size:", matriz_4d.size)
```

Crear una copia de la matriz creada anteriormente.

```python
matriz_3d = matriz_4d[0, :, :, :].copy()
print("Shape 3D:", matriz_3d.shape)
print("Size:", matriz_3d.size)
```

Atributos de la matriz anterior

```python
print("Número de dimensiones:", matriz_3d.ndim)
print("Forma:                ", matriz_3d.shape)
print("Tamaño total:         ", matriz_3d.size)
print("Tipo de dato:         ", matriz_3d.dtype)
print("Bytes por elemento:   ", matriz_3d.itemsize)
print("Bytes totales:        ", matriz_3d.nbytes)
```

Pasar a 2D con reshape

```python
matriz_2d = matriz_3d.reshape(600, 200)
print("Shape 2D:", matriz_2d.shape)
print("Dimensiones:", matriz_2d.ndim)
```

Funcion que convierte la amtriz 2D a DataFrame de Pandas

```python
import numpy as np
import pandas as pd

def matriz_a_dataframe(matriz):
    columnas = [f"col_{i}" for i in range(matriz.shape[1])]
    df = pd.DataFrame(matriz, columns=columnas)
    return df

df = matriz_a_dataframe(matriz_2d)
print("Shape:", df.shape)
print(df.head())
```

Funcion para cargar archivos tipo .mat y .csv

```python
import numpy as np
import pandas as pd
import scipy.io

def cargar_archivo(ruta):
    if ruta.endswith('.csv'):
        datos = pd.read_csv(ruta)
        print("Archivo CSV cargado")
        print("Shape:", datos.shape)
        print(datos.head())
        return datos

    elif ruta.endswith('.mat'):
        datos = scipy.io.loadmat(ruta)
        llaves = [k for k in datos.keys() if not k.startswith('__')]
        print("Archivo mat cargado")
        print("Variables encontradas:", llaves)
        return datos

    else:
        print("Use archivos .mat o .csv")
        return None
```

Funciones de operaciones con NumPy a lo largo de un eje

```python
import numpy as np

def suma(matriz, eje=None):
    return np.sum(matriz, axis=eje)

def resta(m1, m2):
    return np.subtract(m1, m2)

def multiplicacion(m1, m2):
    return np.multiply(m1, m2)

def division(m1, m2):
    return np.divide(m1, m2)

def logaritmo(matriz):
    return np.log(matriz)

def promedio(matriz, eje=None):
    return np.mean(matriz, axis=eje)

def desviacion_estandar(matriz, eje=None):
    return np.std(matriz, axis=eje)

print("Suma total:        ", suma(matriz_2d))
print("Suma por columnas: ", suma(matriz_2d, eje=0))
print("Suma por filas:    ", suma(matriz_2d, eje=1))
print("Promedio total:    ", promedio(matriz_2d))
print("Desv. estándar:    ", desviacion_estandar(matriz_2d))
```

```python
import pandas as pd
import numpy as np

# Cargar el archivo
df_covid = pd.read_csv("covid_19_data.csv")

# Columnas numéricas disponibles
columnas_numericas = ["Confirmed", "Deaths", "Recovered"]

def elegir_columna(mensaje):
    print(f"\nColumnas disponibles: {columnas_numericas}")
    columna = input(mensaje)
    if columna not in columnas_numericas:
        print("Columna no válida")
        return None
    return columna

def suma_pd(df):
    columna = elegir_columna("Elige una columna para sumar: ")
    if columna:
        print(f"Suma de {columna}: {df[columna].sum()}")

def resta_pd(df):
    col1 = elegir_columna("Elige la primera columna: ")
    col2 = elegir_columna("Elige la segunda columna: ")
    if col1 and col2:
        print(f"\nResta {col1} - {col2}:")
        print((df[col1] - df[col2]).head())

def multiplicacion_pd(df):
    col1 = elegir_columna("Elige la primera columna: ")
    col2 = elegir_columna("Elige la segunda columna: ")
    if col1 and col2:
        print(f"\nMultiplicación {col1} * {col2}:")
        print((df[col1] * df[col2]).head())

def division_pd(df):
    col1 = elegir_columna("Elige la primera columna: ")
    col2 = elegir_columna("Elige la segunda columna: ")
    if col1 and col2:
        print(f"\nDivisión {col1} / {col2}:")
        print((df[col1] / df[col2]).head())

def logaritmo_pd(df):
    columna = elegir_columna("Elige una columna para logaritmo: ")
    if columna:
        print(f"\nLogaritmo de {columna}:")
        print(np.log(df[columna].replace(0, np.nan)).head())

def promedio_pd(df):
    columna = elegir_columna("Elige una columna para el promedio: ")
    if columna:
        print(f"Promedio de {columna}: {df[columna].mean()}")

def desviacion_estandar_pd(df):
    columna = elegir_columna("Elige una columna para desviación estándar: ")
    if columna:
        print(f"Desviación estándar de {columna}: {df[columna].std()}")

suma_pd(df_covid)
promedio_pd(df_covid)
desviacion_estandar_pd(df_covid)
```

Graficar una señal tipo .mat

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io

def graficar_mat(ruta):
    datos = scipy.io.loadmat(ruta)
    
    llaves = [k for k in datos.keys() if not k.startswith('__')]
    print("Variables encontradas:", llaves)
    
    # Mostrar las llaves disponibles y elegir
    print("\nLlaves disponibles:")
    for i, llave in enumerate(llaves):
        print(f"{i} - {llave}")
    
    indice = int(input("Elige el número de la variable a graficar: "))
    llave_elegida = llaves[indice]
    señal = datos[llave_elegida].flatten()
    
    fig, ax = plt.subplots()
    ax.plot(señal, label=llave_elegida, color='steelblue')
    ax.set_title(f"Señal: {llave_elegida}")
    ax.set_xlabel("Muestras")
    ax.set_ylabel("Amplitud")
    ax.legend()
    ax.grid(True)
    plt.tight_layout()
    plt.show()
```

Las funciones de graficación piden al usuario títulos, ejes, leyendas y cuadrícula

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_covid = pd.read_csv("covid_19_data.csv")
columnas_numericas = ["Confirmed", "Deaths", "Recovered"]

def pedir_info_grafico():
    titulo   = input("Título del gráfico: ")
    etiqueta_x = input("Etiqueta eje X: ")
    etiqueta_y = input("Etiqueta eje Y: ")
    leyenda  = input("Activar leyenda? (s/n): ")
    cuadricula = input("Activar cuadrícula? (s/n): ")
    return titulo, etiqueta_x, etiqueta_y, leyenda == 's', cuadricula == 's'

def elegir_columna(mensaje):
    print(f"Columnas disponibles: {columnas_numericas}")
    columna = input(mensaje)
    if columna not in columnas_numericas:
        print("Columna no válida")
        return None
    return columna

def graficar_histograma(df):
    columna = elegir_columna("Elige una columna: ")
    if not columna:
        return
    titulo, eje_x, eje_y, leyenda, cuadricula = pedir_info_grafico()

    fig, ax = plt.subplots()
    ax.hist(df[columna], bins=30, label=columna, color='steelblue')
    ax.set_title(titulo)
    ax.set_xlabel(eje_x)
    ax.set_ylabel(eje_y)
    if leyenda:
        ax.legend()
    if cuadricula:
        ax.grid(True)
    plt.tight_layout()
    plt.show()

def graficar_stem(df):
    columna = elegir_columna("Elige una columna: ")
    if not columna:
        return
    titulo, eje_x, eje_y, leyenda, cuadricula = pedir_info_grafico()

    fig, ax = plt.subplots()
    datos = df[columna].head(50)
    ax.stem(datos, label=columna)
    ax.set_title(titulo)
    ax.set_xlabel(eje_x)
    ax.set_ylabel(eje_y)
    if leyenda:
        ax.legend()
    if cuadricula:
        ax.grid(True)
    plt.tight_layout()
    plt.show()

def graficar_barras(df):
    columna = elegir_columna("Elige una columna: ")
    if not columna:
        return
    titulo, eje_x, eje_y, leyenda, cuadricula = pedir_info_grafico()

    fig, ax = plt.subplots()
    datos = df.groupby("Country/Region")[columna].sum().nlargest(10)
    ax.bar(datos.index, datos.values, label=columna, color='coral')
    ax.set_title(titulo)
    ax.set_xlabel(eje_x)
    ax.set_ylabel(eje_y)
    if leyenda:
        ax.legend()
    if cuadricula:
        ax.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

def graficar_pie(df):
    columna = elegir_columna("Elige una columna: ")
    if not columna:
        return
    titulo, eje_x, eje_y, leyenda, cuadricula = pedir_info_grafico()

    fig, ax = plt.subplots()
    datos = df.groupby("Country/Region")[columna].sum().nlargest(7)
    ax.pie(datos.values, labels=datos.index, autopct='%1.1f%%')
    ax.set_title(titulo)
    if leyenda:
        ax.legend()
    plt.tight_layout()
    plt.show()

graficar_histograma(df_covid)
graficar_barras(df_covid)
graficar_pie(df_covid)
graficar_stem(df_covid)
```

Subplots vertical, horizontal o diagonal

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_covid = pd.read_csv("covid_19_data.csv")
columnas_numericas = ["Confirmed", "Deaths", "Recovered"]

def graficar_subplots(df):
    print("\nDiseño de subplots:")
    print("1 - Vertical   (3 filas, 1 columna)")
    print("2 - Horizontal (1 fila, 3 columnas)")
    print("3 - Diagonal   (2x2 usando diagonal)")
    diseno = input("Elige el diseño (1/2/3): ")

    if diseno == '1':
        fig, axes = plt.subplots(3, 1, figsize=(8, 10))
    elif diseno == '2':
        fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    elif diseno == '3':
        fig, axes_2d = plt.subplots(2, 2, figsize=(10, 8))
        axes = [axes_2d[0, 0], axes_2d[1, 1], axes_2d[0, 1]]
        axes_2d[1, 0].set_visible(False)
    else:
        print("Opción no válida")
        return

    titulo_general = input("Título general: ")
    leyenda   = input("Activar leyenda? (s/n): ") == 's'
    cuadricula = input("Activar cuadrícula? (s/n): ") == 's'

    axes[0].hist(df["Confirmed"], bins=30,
                 color='steelblue', label='Confirmed')
    axes[0].set_title("Histograma - Confirmed")
    axes[0].set_xlabel("Casos")
    axes[0].set_ylabel("Frecuencia")
    if leyenda:
        axes[0].legend()
    if cuadricula:
        axes[0].grid(True)

    datos_barras = df.groupby("Country/Region")["Deaths"].sum().nlargest(7)
    axes[1].bar(datos_barras.index, datos_barras.values,
                color='coral', label='Deaths')
    axes[1].set_title("Barras - Deaths por país")
    axes[1].set_xlabel("País")
    axes[1].set_ylabel("Muertes")
    axes[1].tick_params(axis='x', rotation=45)
    if leyenda:
        axes[1].legend()
    if cuadricula:
        axes[1].grid(True)

    axes[2].plot(df["Recovered"].head(100),
                 color='green', label='Recovered')
    axes[2].set_title("Línea - Recovered")
    axes[2].set_xlabel("Índice")
    axes[2].set_ylabel("Recuperados")
    if leyenda:
        axes[2].legend()
    if cuadricula:
        axes[2].grid(True)

    fig.suptitle(titulo_general, fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()

graficar_subplots(df_covid)
```
